```python
import matplotlib.pyplot as plt
import numpy as py 
from sklearn import datasets, linear_model, model_selection

# Load in the dataset
X, y = datasets.load_diabetes(return_X_y=True)

# Print the shape of the data and the first row
print(X.shape)
print(X[0])
```

```python
# Extract the first (and only) column
X = X[:, 2]
print(X.shape)

# Reshape to 2D array
X = X.reshape((-1,1))
print(X.shape)
```

```python
# Split the data into training and testing data
X_train, X_test, y_train, y_test, = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
# Create the linear regression model
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict using out test data
y_pred = model.predict(X_test)
```

# How did we do? Check with Matplot lib

```python
# Create a scatter plot
plt.scatter(X_test, y_test, color='black')

# Plot the predictions
plt.plot(X_test, y_pred, color='blue', linewidth=3)

# Format the scatter plot
plt.xlabel('Scaled BMI\'s')
plt.ylabel('Disease Progression')
plt.title('Diabetes Progression against BMI')

# Print
plt.show()
```
