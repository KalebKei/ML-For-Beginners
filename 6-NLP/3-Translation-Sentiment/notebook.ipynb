```python
from textblob import TextBlob

quote1 = """It is a truth universally acknowledged, that a single man in possession of a good fortune, must be in want of a wife."""

quote2 = """Darcy, as well as Elizabeth, really loved them; and they were both ever sensible of the warmest gratitude towards the persons who, by bringing her into Derbyshire, had been the means of uniting them."""

sentiment1 = TextBlob(quote1).sentiment
sentiment2 = TextBlob(quote2).sentiment

print(quote1 + " has a sentiment of " + str(sentiment1))
print(quote2 + " has a sentiment of " + str(sentiment2))
```

```python
file = open("Pride_and_Prejudice.txt")
# print(file.readline())

sentence = file.readline()
wompWomp = []
good = []
while(sentence):
    sentiment = TextBlob(sentence).sentiment
    # print(sentence + " has a sentiment of " + str(sentiment))
    if(sentiment.polarity < 0.0):
        wompWomp.append(sentence)
    elif(sentiment.polarity >= 0.0):
        good.append(sentence)
    sentence = file.readline()

print("Length of positive sentiment list: " + str(len(good)))
print("Length of negative sentiment list: " + str(len(wompWomp)))

print("Here are the < 0 entries")
for x in wompWomp:
    print(x + '\n')

print("Here are the >= entries")
for x in good:
    print(x + "\n")
```
